```python
# import and read file
import pandas as pd
fd = flight_data = pd.read_csv("Combined Flights Data/combined_flights_2019.csv")

fd.head()
```

```python
# filtering columns
columns = ['callsign', 'origin', 'destination', 'day']
ffd = filteres_flights_data = flight_data[columns]

ffd.head()
```

```python
# select rows that contains netherlands airport
od = origin_or_destination = ffd.loc[((ffd['origin'] == 'EHAM') | 
                                      (ffd['origin'] == 'EHEH') | 
                                      (ffd['origin'] == 'EHGG') | 
                                      (ffd['origin'] == 'EHBK') | 
                                      (ffd['origin'] == 'EHRD')) | 
                                     ((ffd['destination'] == 'EHAM') | 
                                      (ffd['destination'] == 'EHEH') | 
                                      (ffd['destination'] == 'EHGG') | 
                                      (ffd['destination'] == 'EHBK') | 
                                      (ffd['destination'] == 'EHRD'))]

od
```

```python
# remove missing values
clean_od = od.dropna()

clean_od
```

```python
# taking airline code only
clean_od['airline'] = clean_od['callsign'].str[:3]

# reset column composition
columns = ['origin', 'destination', 'day', 'airline']
tcnf = total_clean_NL_flight = clean_od[columns]

tcnf
```

```python
# export to csv
tcnf.to_csv("all_NL_flights_2019_airlines.csv",
                    index = False,
                    encoding = 'utf-8-sig')
```
